```python
import sqlite3
import csv
from sqlite3 import Error
from sqlite3 import connect
 

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return None
        
def get_data(conn):

    with open('data_with_imgurl2.csv', errors='ignore') as csvfile:
        reader = csv.reader(csvfile)
        contents = [row for row in reader]
    print('data fetched')
    print('contents list length: ' + str(len(contents)))
    
    for i in range(1, len(contents)):
        add_content(conn, contents[i])
        print('adding ' + str(i))
    print('data added to database')

def create_table(conn, table_query):
    try:
        c = conn.cursor()
        c.execute(table_query)
    except Error as e:
        print(e)
        
def add_content(conn, lst):
    curr = conn.cursor()
    query = 'INSERT INTO posts( \
                post_id, \
                handle, \
                postDate, \
                postType, \
                gen_postType,\
                postUrl, \
                platform, \
                message, \
                twitter_comment, \
                twitter_retweet, \
                relevant_prediction, \
                cta_prediction, \
                sell_prediction, \
                intent_prediction, \
                comments, \
                shares, \
                likes, \
                views, \
                campaignId, \
                campaign_name, \
                campaign_startDate, \
                campaign_endDate, \
                project_currency, \
                project_staffPick, \
                sub_category, \
                category_names, \
                category_numbers, \
                project_url, \
                project_backerCount, \
                project_goal, \
                project_pledgeAmount, \
                succeeded, \
                imageUrl) VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)'
    curr.execute(query,lst[1:])
    
def main():
    database = 'database.db'
    table_query = 'CREATE TABLE \
                IF NOT EXISTS posts(post_id text NOT NULL, \
                handle text NOT NULL, \
                postDate text NOT NULL, \
                postType text NOT NULL, \
                gen_postType text NOT NULL,\
                postUrl text NOT NULL, \
                platform text NOT NULL, \
                message text, \
                twitter_comment text NOT NULL, \
                twitter_retweet text NOT NULL, \
                relevant_prediction text NOT NULL, \
                cta_prediction text NOT NULL, \
                sell_prediction text NOT NULL, \
                intent_prediction text NOT NULL, \
                comments text NOT NULL, \
                shares text NOT NULL, \
                likes text NOT NULL, \
                views text NOT NULL, \
                CampaignId text NOT NULL, \
                campaign_name text NOT NULL, \
                campaign_starteDate text NOT NULL, \
                campaign_endDate text NOT NULL, \
                project_currency text NOT NULL, \
                project_staffPick text NOT NULL, \
                sub_category text NOT NULL, \
                CategoryNames text NOT NULL, \
                CategoryNumbers text NOT NULL, \
                project_url text NOT NULL, \
                project_backerCount text NOT NULL, \
                project_goal text NOT NULL, \
                project_pledgeAmount text NOT NULL, \
                Succeeded text NOT NULL, \
                imageUrl text NOT NULL \
               );'

    conn = create_connection(database)
    print('connection made')
    if conn is not None:
        print('creating table')
        create_table(conn, table_query)#create projects table
    else:
        print("Error")
    print('fetching data')
    with conn:
        get_data(conn)

if __name__ == '__main__':
    main()
    
```
