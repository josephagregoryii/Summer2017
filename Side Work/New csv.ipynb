```python
import csv

def get_data():
    
    with open('Large_Data_Set_Posts15.csv', errors ='ignore') as csvfileLarge:
        reader1 = csv.reader(csvfileLarge)
        mainContents = [row for row in reader1]
    tables = mainContents[0]
    tables.append('image_URL')
    
    with open('LargeData16.csv', errors ='ignore') as csvfileSingle:
        reader2 = csv.reader(csvfileSingle)
        subContents = [row for row in reader2]
            
    results = []
    for cell in mainContents[1:]:
        message = cell[8].lower()
        intent = cell[14]
        if message.find("rt") != -1 or intent == '5':
            dict = { 
                'post_id' : cell[1],
                'handle' : cell[2], 
                'postDate' : cell[3], 
                'postType': cell[4], 
                'gen_postType': cell[5], 
                'postUrl': cell[6], 
                'platform': cell[7], 
                'message': cell[8], 
                #'twitter_comment' : cell[9], 
                #'twitter_retweet': cell[10], 
                'relevant_prediction': cell[11], 
                'cta_prediction': cell[12], 
                'sell_prediction': cell[13], 
                'intent_prediction': cell[14], 
                'comments': cell[15], 
                'shares': cell[16], 
                'likes': cell[17], 
                'views': cell[18], 
                'CampaignId': cell[19], 
                'campaign_name': cell[20], 
                'campaign_startDate': cell[21], 
                'campaign_endDate': cell[22], 
                'project_currency': cell[23], 
                'project_staffPick': cell[24], 
                'sub_category': cell[25], 
                'project_url': cell[26], 
                'project_backerCount': cell[27], 
                'project_goal': cell[28], 
                'project_pledgeAmount': cell[29], 
                'Succeeded': cell[30]
            }
            results.append(dict)
    
    image_URLS = []
    for cell2 in subContents:
        if cell2[5] != '':
            dict = {
                "post_id" : cell2[0],
                "image_URL" : cell2[5]
            }
            image_URLS.append(dict)

    match = []
    for j in image_URLS:
        for i in results:
            if j['post_id'] == i["post_id"]:
                i['image_URL'] = j['image_URL']
                match.append(i)
            else:
                i['image_URL'] = ''

    with open('newfile.csv', 'w') as csvfile:
        fieldnames = ['first_name', 'last_name']
        writer = csv.DictWriter(csvfile, fieldnames=tables)
        writer.writeheader()
        for cell in results:
            writer.writerow({
                'post_id' : cell['post_id'],
                'handle' : cell['handle'], 
                'postDate' : cell['postDate'], 
                'postType': cell['postType'], 
                'gen_postType': cell["gen_postType"], 
                'postUrl': cell['postUrl'], 
                'platform': cell["platform"], 
                'message': cell['message'], 
                'relevant_prediction': cell["relevant_prediction"], 
                'cta_prediction': cell["cta_prediction"], 
                'sell_prediction': cell['sell_prediction'], 
                'intent_prediction': cell['intent_prediction'], 
                'comments': cell['comments'], 
                'shares': cell['shares'], 
                'likes': cell['likes'], 
                'views': cell['views'], 
                'CampaignId': cell['CampaignId'], 
                'campaign_name': cell['campaign_name'], 
                'campaign_startDate': cell['campaign_startDate'], 
                'campaign_endDate': cell['campaign_endDate'], 
                'project_currency': cell['project_currency'], 
                'project_staffPick': cell['project_staffPick'], 
                'sub_category': cell["sub_category"], 
                'project_url': cell['project_url'], 
                'project_backerCount': cell['project_backerCount'], 
                'project_goal': cell['project_goal'], 
                'project_pledgeAmount': cell['project_pledgeAmount'], 
                'Succeeded': cell['Succeeded'],
                'image_URL': cell['image_URL']
            })
    
```
