```python
import csv
import pandas as pd

def openfile(filename):
    name_list = []
    scraper_id = []
    graph_type = []

    with open(filename, encoding = 'ascii', errors = 'ignore') as csvfile:
        reader = csv.reader(csvfile)
        content2 = [row for row in reader]
        csvfile.close()
    table2 = content2[4]

    prev2 = ''
    for cell in content2[1:]:
        temp2 = []
        ID = cell[2]
        name = cell[0]
        graph = cell[4]
        if ID == '':
            cell[2] = 'N/A'
            ID = cell[2]
            name_list.append(name)
            scraper_id.append(ID)
            graph_type.append(graph)
            prev = ID
        elif graph == '':
            cell[4] = 'Type 0 0/0'
            graph = cell[4]
            name_list.append(name)
            scraper_id.append(ID)
            graph_type.append(graph)
            prev = ID
        elif ID != prev2:
            name_list.append(name)
            scraper_id.append(ID)
            graph_type.append(graph)
            prev = ID
    return [scraper_id, graph_type]

def parseList(IDList, graphList):
    bulk = []
    answerList = []
    for i in range(0, len(IDList)):
        answerList.append(IDList[i])
        dic = parseGraphType(graphList[i])
        answerList.append(dic['type'])
        answerList.append(dic['funded'])
        answerList.append(dic['reachedAt'])
        bulk.append(answerList)
        answerList = []
    return bulk

def parseList2(IDList, graphList):
    bulk = []
    typeList = []
    fundedList = []
    reachedList = []

    for i in range(0, len(IDList)):
        #answerList.append(IDList[i])
        dic = parseGraphType(graphList[i])
        typeList.append(dic['type'])
        fundedList.append(dic['funded'])
        reachedList.append(dic['reachedAt'])
    return [typeList, fundedList, reachedList]

def parseGraphType(input):
    list = input.split() #split by whitespace
    dictionary = {'type': '0', 'funded': '0', 'reachedAt': '0'}
    if list[0] != "Type":
        return dictionary
    else:
        dictionary['type'] = list[1]
        if len(list) < 3:
            return {'type': '0', 'funded': '0', 'reachedAt': '0'}
        else:
            if "/" in list[2]:
                numbers = list[2].split('/')
                if dictionary['type'] == '3':
                    dictionary['funded'] = numbers[0]
                    dictionary['reachedAt'] = '0'
                    return dictionary
                else:
                    dictionary['funded'] = numbers[0]
                    dictionary['reachedAt'] = numbers[1]
                    return dictionary
            else:
                numbers = list[2]
                if dictionary['type'] == '3':
                    dictionary['funded'] = numbers
                    dictionary['reachedAt'] = '0'
                    return dictionary
                else:
                    return {'type': '0', 'funded': '0', 'reachedAt': '0'}

def listOfLists(fileName):
    lists = openfile('preFinal.csv')
    IDList = lists[0]
    graphList = lists[1]
    output = parseList2(IDList, graphList)
    return output

def outputcsv(typeList,fundedList,reachedList):
    f = open('output.csv','w')
    writer = csv.writer(f)
    writer.writerow( ('type','funded','reached') )
    for i,j,z in zip(typeList,fundedList,reachedList):
        f.write(str(i)+","+str(j)+","+str(z)+"\n")
    f.close()
 
"""" 
def combinecsv():
    to_merge = ['ScraperCampaignsNew.csv','output.csv']
    dfs = []
    for filename in to_merge:
        df = pd.read_csv(filename, header=None, encoding = 'latin1',skip_blank_lines = False)
        df.columns = [filename + str(cname) for cname in df.columns]
        dfs.append(df)

    merged = pd.concat(dfs,axis=1)
    merged.to_csv("merged_data.csv", header=None, index=None)
    """

def main():
    output = listOfLists('preFinal.csv')
    outputcsv(output[0],output[1],output[2])
    #combinecsv()
    

if __name__ == '__main__':
    main()
```
